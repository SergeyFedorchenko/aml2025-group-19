```python
import os
import glob
from cleanfid import fid
from PIL import Image
import numpy as np
from tqdm import tqdm

# Directory with your mixed images
image_dir = r"C:\Users\sergei\train_gan\pytorch-CycleGAN-and-pix2pix\checkpoints\cartoon_transfer_prost_loss\web\images"  # replace with your path
all_files = sorted(os.listdir(image_dir))

# Group by epoch
epochs = sorted(set(f.split("_")[0] for f in all_files if f.endswith(".png")))

# Optional: restrict to fake_B vs real_B
target_fake = "fake_B"
target_real = "real_B"

fid_per_epoch = {}

for epoch in tqdm(epochs):
    fake_imgs = sorted([f for f in all_files if f.startswith(epoch) and target_fake in f])
    real_imgs = sorted([f for f in all_files if f.startswith(epoch) and target_real in f])
    
    if not fake_imgs or not real_imgs:
        continue
    
    # Load images into numpy arrays
    def load_imgs(img_list):
        arrs = []
        for name in img_list:
            img = Image.open(os.path.join(image_dir, name)).convert("RGB").resize((299, 299))
            arr = np.array(img).astype(np.uint8)
            arrs.append(arr)
        return np.stack(arrs)

    imgs_fake = load_imgs(fake_imgs)
    imgs_real = load_imgs(real_imgs)

    # Compute FID using cleanfid (on raw numpy arrays)
    fid_val = fid.compute_fid(
        samples1=imgs_real, 
        samples2=imgs_fake,
        mode="clean"  # or "legacy" for compatibility
    )
    fid_per_epoch[epoch] = fid_val

# Print or plot results
for epoch, val in fid_per_epoch.items():
    print(f"{epoch}: FID = {val:.2f}")
```

```python
import os
import torch
from torchvision import transforms
from torchvision.utils import save_image
from PIL import Image

# --- CONFIG ---
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
MODEL_PATH = r"C:\Users\sergei\train_gan\pytorch-CycleGAN-and-pix2pix\checkpoints\cartoon_transfer_prost_percep\latest_net_G_A.pth"  # path to your trained generator
INPUT_FOLDER = "input_photos"   # folder with test images
OUTPUT_FOLDER = "output_cartoon"  # folder to save cartoonified images
IMG_SIZE = 256

# --- LOAD MODEL ---
from models.networks import define_G  # assuming standard CycleGAN structure

#netG_A = define_G(input_nc=3, output_nc=3, ngf=64, netG='resnet_9blocks', norm='instance', use_dropout=False, init_type='normal', 
#                  init_gain=0.02, gpu_ids=[0] if DEVICE=="cuda" else [])
netG_A = define_G(
    input_nc=3, output_nc=3, ngf=64, netG='resnet_9blocks',
    norm='instance', use_dropout=False, init_type='normal',
    init_gain=0.02, gpu_ids=[0] if DEVICE == "cuda" else []
)
state_dict = torch.load(MODEL_PATH, map_location=DEVICE)
clean_state_dict = {k.replace("module.", ""): v for k, v in state_dict.items()}

netG_A.load_state_dict(clean_state_dict)

netG_A.to(DEVICE)
netG_A.eval()
#netG_A.load_state_dict(torch.load(MODEL_PATH, map_location=DEVICE))
#netG_A.to(DEVICE)
#netG_A.eval()

# --- TRANSFORM ---
transform = transforms.Compose([
    transforms.Resize(IMG_SIZE),
    transforms.CenterCrop(IMG_SIZE),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

# --- INFERENCE FUNCTION ---
def run_inference(img_path, out_path):
    img = Image.open(img_path).convert("RGB")
    input_tensor = transform(img).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        output = netG_A(input_tensor)

    output = (output + 1) / 2  # [-1,1]  [0,1]
    save_image(output, out_path)

# --- MAIN LOOP ---
os.makedirs(OUTPUT_FOLDER, exist_ok=True)
for fname in os.listdir(INPUT_FOLDER):
    if fname.lower().endswith((".png", ".jpg", ".jpeg")):
        in_path = os.path.join(INPUT_FOLDER, fname)
        out_path = os.path.join(OUTPUT_FOLDER, fname)
        run_inference(in_path, out_path)
        print(f"Saved: {out_path}")
```
